```
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


globalData = np.loadtxt('/content/drive/Shared drives/ASA DataFest 2020/Code and Graphs/Raw Data/Global.csv', delimiter = ',', skiprows = 1)
usData = np.loadtxt('/content/drive/Shared drives/ASA DataFest 2020/Code and Graphs/Raw Data/United States.csv', delimiter = ',', skiprows = 1)

#The data consists of 21 weeks of material
week = globalData[:,0]

#MVB is the percent change in Mobile download speeds Versus the Base week
#The base week was designated as the 12th week, March 2nd to 8th, about two weeks before stay-at-home orders were initiated in many states

globalmvb = globalData[:,2] * 100
usmvb = usData[:,2] * 100

#MAD is the Mobile Average Download speed

globalmad = globalData[:,3]
usmad = usData[:,3]

#FVB is the percent change in Fixed download speeds Versus the Base week

globalfvb = globalData[:,5] * 100
usfvb = usData[:,5] * 100

#FAD is the Fixed Average Download speed

globalfad = globalData[:,6]
usfad = usData[:,6]
```

```
plt.plot(week, globalmvb, label = "Global", color = 'teal')
plt.plot(week, usmvb, label = "United States", color = 'orange')
plt.hlines(0, 0, 20, label = 'Baseline', color = 'black')
plt.vlines(12, -12.5, 2.5, label = "Base Week", color = 'grey')
plt.ylabel("% Change in Mobile Download Speeds")
plt.xlabel("Week #, Starting from Monday 12/16/19")
plt.title("Mobile Download Speeds Per Week Relative to Base Week (3/8/20)")
plt.legend()
plt.savefig('graph1.png', dpi=1920)
```

```
plt.plot(week, globalfvb, label = "Global", color = 'teal')
plt.plot(week, usfvb, label = "United States", color = 'orange')
plt.hlines(0, 0, 20, label = 'Baseline', color = 'black')
plt.vlines(12, -10, 1, label = "Base Week", color = 'grey')
plt.ylabel("% Change in Fixed Download Speeds")
plt.xlabel("Week #, Starting from Monday 12/16/19")
plt.title("Fixed Download Speeds Per Week Relative to Base Week (3/8/20)")
plt.legend()
plt.savefig('graph2.png', dpi=1920)
```

```
#Extra Graph Not Used in Material

plt.plot(week, globalfad, label = "Global", color = 'teal')
plt.plot(week, usfad, label = "United States", color = 'orange')
plt.ylabel("Average Download Speeds")
plt.xlabel("Week #, Starting from Monday 12/16/19")
plt.title("Fixed Download Speed: Comparison Between Global and US Averages")
plt.legend()
plt.savefig('graph3.png', dpi=1920)
```
