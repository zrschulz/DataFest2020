```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```
data = np.loadtxt('/content/drive/My Drive/Datafest/CSVs/JanMayPrices.csv', delimiter = ',')
data2 = np.loadtxt('/content/drive/My Drive/Datafest/CSVs/Stock Prices - Indexes.csv', delimiter = ',')
```

```
#arrays of prices for each company
verizon = data[:,0]
atnt = data[:,1]
frontier = data[:,2]
cox = data[:,3]
ctl = data[:,4]
charter = data[:,5]
comcast = data[:,6]
tmobile= data[:,7]
wow = data[:,8]
windstream = data[:,9]



#scale spectrum data to fit frame of graph
charter = charter * (1/5)



#array for x-axis
time = np.arange(0,90,1)

#plt.plot each company's share prices over time
plt.plot(time,verizon, label = 'Verizon', )
plt.plot(time,atnt, label = 'AT&T' )
plt.plot(time,ctl, label = 'CenturyLink' )
plt.plot(time,comcast, label = 'Comcast')
plt.plot(time,tmobile, label = 'TMobile' )
plt.plot(time,charter, label = 'Spectrum (1/5 of actual value)' )




#mark date of first statewide lockdown with vertical line
plt.vlines([55], 0, 110, color = 'green', linewidth = 3, linestyle = '--', label = 'First Statewide Lockdown (California)')



#mark dates on x-axis
plt.xticks(ticks=[0,22,41,63,84], labels=[ '(1/2/20)','(2/3/20)','(3/2/20)','(4/1/20)','(5/1/20)'])

#axis labels
plt.xlabel("Date")
plt.ylabel("Share Price (USD)")

#create legend
plt.legend(loc = 'lower right')

#title
plt.title("Internet Service Provider Stock Prices in 2020")

#resize
fig = plt.gcf()
fig.set_size_inches(12, 9)
plt.savefig('graph1F2.png', dpi=1920)
```
