```
import numpy as np #import numpy
import matplotlib.pyplot as plt #import matlab ploting library
```

```
N = 4 #number of dates
ind = np.arange(N)
width = 0.2 #width of each bin in each date category
gaming = [0.75, 1.02, 0.04, -0.01] #gaming percentages
VPN = [0.34, 0.40, -0.05, 0.05] #VPN traffic percentages
web = [0.20, 0.24, -0.02, 0.02] #web traffic percentages
video = [0.12, 0.33, 0.00, 0.00] #video percentages
space = [0.00, 0.00, 0.00, 0.00] #creates a space

X = np.arange(4) #four date bins
fig = plt.figure() #plot figure
ax = fig.add_axes([0.09,0.09,.85,.85])
ax.bar(X + 0.00, gaming, color = 'b', label = 'Gaming', width = 0.2) #gaming percentage at each date
ax.bar(X + 0.20, VPN, color = 'g', label = 'VPN Traffic', width = 0.2) #VPN traffic percentage at each date
ax.bar(X + 0.40, web, color = 'r', label = 'Web Traffic', width = 0.2) #web traffic percentage at each date
ax.bar(X + 0.60, video, color = 'c', label = 'Video', width = 0.2) #video percentage at each date
ax.bar(X + 0.80, space, color = 'w', width = 0.2) #space
ax.set_xticks(ind+width) #places the date labels
ax.set_xticklabels(('17 March 2020', '02 April 2020', '09 April 2020', '22 April 2020')) #date labels
plt.legend(loc = 'best') #legend 
plt.title('Week over Week Change for Internet Usage') #title
plt.ylabel('Percent Change') #y-axis title

plt.savefig('graph1.png', dpi=1920)
```
